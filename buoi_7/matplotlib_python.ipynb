```python
df = pd.read_csv("/Users/vietannguyen/datasets/cleaned_data.csv")
df.info()
```

### Column chart (biểu đồ cột)

```python
# Bước 1: tổng hợp dữ liệu 
purchase_amount_by_items = df.groupby("Item Purchased").agg({"Purchase Amount (USD)": "sum"})
purchase_amount_by_items
```

```python
# Bước 2: Vẽ biểu đồ cột 

# Xác định kích thước của biểu đồ
plt.figure(figsize=(25, 6)) 
# Vẽ biểu đồ 
# .bar(giá trị hiển thị ở trục hoành, giá trị hiển thị trên trục tung)
plt.bar(purchase_amount_by_items.index, purchase_amount_by_items["Purchase Amount (USD)"])
plt.title("Columns chart")
plt.xlabel("Items")
plt.ylabel("Purchase Amount")
```

### Biểu đồ tròn

```python
payment_method_by_customer = df.groupby("Payment Method").agg({"Customer ID": "count"})
payment_method_by_customer
```

```python
plt.figure(figsize=(15, 6))
# .pie(số lượng, label) 
plt.pie(payment_method_by_customer["Customer ID"], labels=payment_method_by_customer.index, autopct="%1.1f%%")
plt.title("Pie chart")
plt.show()
```

### Biểu đồ phân tán

```python
purchase_amount_by_age = df.groupby("Age").agg({"Purchase Amount (USD)": "sum"})
purchase_amount_by_age 
```

```python
plt.figure(figsize=(25, 6)) 
# Vẽ biểu đồ 
# .bar(giá trị hiển thị ở trục hoành, giá trị hiển thị trên trục tung)
plt.scatter(purchase_amount_by_age.index, purchase_amount_by_age["Purchase Amount (USD)"])
plt.title("Scatter chart")
plt.xlabel("Age")
plt.ylabel("Purchase Amount")
```

### Biểu đồ thanh ngang

```python
# Bước 2: Vẽ biểu đồ cột 
# Sort lại giá trị 
top_10 = purchase_amount_by_items.sort_values(by="Purchase Amount (USD)", ascending=True).head(10)
# Xác định kích thước của biểu đồ
plt.figure(figsize=(25, 6)) 
# Vẽ biểu đồ 
# .bar(giá trị hiển thị ở trục hoành, giá trị hiển thị trên trục tung)
plt.barh(top_10.index, top_10["Purchase Amount (USD)"])
plt.title("Columns chart")
plt.xlabel("Items")
plt.ylabel("Purchase Amount")
```

### **Biểu đồ cột**

```python
plt.figure(figsize=(25, 6))
# Tham số: trục x (hoành), trục y (trục tung), color 
sns.barplot(x=purchase_amount_by_items.index, y=purchase_amount_by_items["Purchase Amount (USD)"], color="green")
```

```python
plt.figure(figsize=(25, 6))
sns.scatterplot(x = purchase_amount_by_age.index, y = purchase_amount_by_age["Purchase Amount (USD)"])
```

### **Biểu đồ thanh ngang**

```python
plt.figure(figsize=(25, 6))
top_10 = purchase_amount_by_items.sort_values(by="Purchase Amount (USD)", ascending=False).head(10)
sns.barplot(y = top_10.index, x = top_10["Purchase Amount (USD)"])
```
