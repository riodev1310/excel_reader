```python
df_expensive = df[df["price"] > 300] 
df_expensive.head() 
```

```python
# price dưới 500 
con1 = df["price"] < 500 

# price trên 200 
con2 = df["price"] > 200 

# Số lượng bán ra dưới 10
con3 = df["amount"] < 10

# Tổng thu nhập lớn hơn 50000
con4 = df["total"] > 50000
```

```python
# Sản phẩm rẻ và số sale thấp
df[con1 & con3]
```

```python
# Sản phẩm hoặc là rất rẻ hoặc là rất mắc
df[con1 | con2]
```

```python
# Sản phẩm với giá 200 trở xuống
df[~con2] 
```

```python
# Sản phẩm rẻ hoặc số lượng bán ra thấp, nhưng lại có thu nhập cao
df[(con1 | con2) & con4]
```
